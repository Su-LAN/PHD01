```python
from pathlib import Path
import json
import pandas as pd

candidates = [
    Path("ddxplus_test_results_llama3.1_8b.json"),
]
results_path = next((p for p in candidates if p.exists()), None)
if results_path is None:
    raise FileNotFoundError(f"找不到结果文件，已尝试: {', '.join(str(p) for p in candidates)}")

records = json.loads(results_path.read_text(encoding="utf-8"))
df = pd.DataFrame(records)

def to_bool(x):
    if isinstance(x, bool):
        return x
    if x is None:
        return False
    s = str(x).strip().lower()
    if s in {"true", "1", "yes", "y", "t"}:
        return True
    if s in {"false", "0", "no", "n", "f"}:
        return False
    return False

if "is_correct" not in df.columns:
    raise KeyError("JSON中缺少字段: is_correct")

df["is_correct"] = df["is_correct"].map(to_bool)
df["has_error"] = df["error"].fillna("").astype(str).str.len().gt(0) if "error" in df.columns else False

total = len(df)
correct = int(df["is_correct"].sum())
errors = int(df["has_error"].sum())
accuracy = correct / total if total else 0.0

print(f"Overall accuracy: {correct}/{total} = {accuracy:.2%} (errors={errors})")

```
